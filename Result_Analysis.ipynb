```python
import numpy as np 
import pandas as pd 
import requests
import io
```

```python
# Downloading the result data/file from your GitHub account

url_1 = ("https://raw.githubusercontent.com/Kensuzuki95/Corporate_AI_Ethics_Guideline_Analysis/main/Result_Data/Similarity_Analysis_Results.csv")
download = requests.get(url_1).content

results = pd.read_csv(io.StringIO(download.decode('utf-8')))

results.info()
results.head()
```

```python
# Create a column showing column name with Maximum and Minimum values of each row
results['Max_Principle'] = results.iloc[:, 1:10].idxmax(axis=1)
results['Min_Principle'] = results.iloc[:, 1:10].idxmin(axis=1)

# Create a column showing the Maximum and Minimum value of each rows
results['Max_Val'] = results.iloc[:, 1:10].max(axis=1)
results['Min_Val'] = results.iloc[:, 1:10].min(axis=1)
results.head()
```

```python
results['Max-Min'] = results.iloc[:, 1:10].max(axis=1) - results.iloc[:, 1:10].min(axis=1)
results['Avg'] = results.iloc[:, 1:10].mean(axis=1)
results.head()
```
